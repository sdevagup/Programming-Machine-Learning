#### 11.Now consider a modified version of the bank data that contains missing values (bank_data_missing_vals.csv). (a) Using Pandas determine all the attributes with missing values and the number of missing values for each such attribute. (b) Show all the instances in the data that contain a missing value. (c) Fill the missing values for all numeric attributes using the mean value for the attribute. (d) After filling in the missing numeric values, drop all rows where a categorical attribute contains a missing value. (e) Show that the final resulting table does not contain missing values.

```python
df_miss=pd.read_csv("http://facweb.cs.depaul.edu/mobasher/classes/CSC478/Assignments/bank_data_missing_vals.csv", sep=',')
df_miss.info()
```

```python
# checking for non integer values in numercial columns
list1 =["age","income","children"]
for i in list1:
    mask = pd.to_numeric(df_miss[i], errors='coerce').isna()
    print(i+" = "+str(mask.sum()))
```

```python
# checking for unique values in catergorical columns
list2 = ["gender","region","married","car","savings_acct","current_acct","mortgage","pep"]
for i in list2:
    print(df_miss[i].value_counts())
```

```python
#replacing the '?' with nan
df_miss = df_miss.replace(['?'],np.nan)
```

```python
# calculating sum of null values
df_miss.isnull().sum()
```

```python
# replacing null values with mean of the column
df_miss["age"]=df_miss["age"].astype(float)
df_miss["age"] = df_miss["age"].replace(np.nan,df_miss["age"].mean())
```

```python
# replacing null values with mean of the column
df_miss["income"]=df_miss["income"].astype(float)
df_miss["income"] = df_miss["income"].replace(np.nan,df_miss["income"].mean())
```

```python
# droping categorical columns with null values 
df_miss = df_miss.drop(columns=['gender','region'])
```

```python
#checking for null values
df_miss.isnull().sum()
```

The given file contains cells with symbol '?' which represents null values. So i replaced the ? with null values and calculated the sum of null values for each row. Then, replaced numerical columns with null values with mean and droped  the categorical colums
